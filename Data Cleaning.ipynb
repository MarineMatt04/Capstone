```python
# Assigning a value to my second data set:

movie_data = pd.read_csv(r'C:\Users\mattl\code\SavvyCoders\Capstone\Data Sets\CumulativeMovieData.csv')
movie_data
```

```python
# Initial cleaning of second data set by removing unnecessary columns:

movie_data.drop(['rating', 'score', 'votes', 'director', 'writer', 'star', 'country', 'budget', 'company', 'runtime', 'released'], axis=1, inplace=True)
```

```python
# Identifyig 'Released' date format:

def parse_date(date_string):
    try:
        date_string = date_string.split('(')[0].strip()
    except AttributeError:
        print("Error with date1: " + str(date_string))
        return None
    try:
        date_object = datetime.strptime(date_string, "%B %d, %Y")
    except ValueError:
        try:
            date_object = datetime.strptime(date_string, "%B %Y")
        except ValueError:
            try:
                date_object = datetime.strptime(date_string, "%Y")
            except ValueError:
                print("Error with date: " + date_string)
                return None
            date_object = datetime.strptime(date_string, "%Y")
    return date_object.strftime("%Y-%m-%d")

# Appling the function to the 'date' column:

movie_data['released2'] = movie_data['Cleaned_released_date'].apply(parse_date)

# Converting 'released2' to appropriate datetime format:

movie_data['released2'] = pd.to_datetime(movie_data['released2'], errors='coerce')
movie_data
```

```python
# Removing initial 'date' column:

movie_data.drop(['Cleaned_released_date'], axis=1, inplace=True)
```

```python
# Testing implementation of code and date format:

movie_data = movie_data[movie_data['released2']!='NaT']
movie_data
```

```python
# Trimming my data to a specific date range:

startDate = pd.to_datetime('2014-01-01')
endDate = pd.to_datetime('2019-12-31')
smMovieData = movie_data[(movie_data['released2'] >= startDate) & (movie_data['released2'] <= endDate)]
smMovieData
```

```python
# Sorting my 'date' range in descending order to match style of first data set:

smMovieData.sort_values(by = ['released2'], inplace=True, ascending=False)
smMovieData
```

```python
# Saving my second data set to be accessed through other platforms:

main_csv_file_path = r'C:\Users\mattl\code\SavvyCoders\Capstone\Data Sets\Cumulative Movie Data.csv'

smMovieData.to_csv(main_csv_file_path, index=False)

print(f"DataFrame saved to {main_csv_file_path}")
```
